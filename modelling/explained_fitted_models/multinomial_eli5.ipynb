In this jupyter notebook are logistic regression, desigion tree, random forest and xgboost models fitted to black friday dataset and explained by eli5, lime and shap.

```python
import eli5
import lime

import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, classification_report
```

## Data loading

```python
BLACK_FRIDAY_PATH = "../../datasets/preprocessed/pre_black_friday.csv"
```

```python
black_df = pd.read_csv(BLACK_FRIDAY_PATH) 
black_df.head()
```

```python
X = black_df.drop(["Unnamed: 0","Purchase_x"], axis=1)
Y = black_df["Purchase_x"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.3, random_state=42)
```

## LogisticRegression

```python
# models = {
#     "Logistic Regression": LogisticRegression(C=1e5, solver='lbfgs', multi_class='multinomial', max_iter=10000),
#     "Decision Tree": DecisionTreeClassifier(),
#     "Random Forest Classifier": RandomForestClassifier(n_estimators=10),
#     "XGboost": XGBClassifier()
# }

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
feature_names = list(X_train.columns)
```

```python
eli5.show_weights(lr, feature_names=feature_names) 
```

```python
eli5.show_prediction(lr, X_test.iloc[0])
```

## DecisionTreeClassifier

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
eli5.show_weights(dtc, feature_names=feature_names, show=['feature_importances']) 
```

```python
eli5.show_prediction(dtc, X_test.iloc[0])
```

## RandomForestClassifier

```python
rfc = RandomForestClassifier(n_estimators=10)
rfc.fit(X_train, y_train)
```

```python
eli5.show_weights(rfc, feature_names=feature_names, show=['feature_importances']) 
```

```python
eli5.show_prediction(rfc, X_test.iloc[0])
```

## XGboost

```python
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
eli5.show_weights(xgb, feature_names=feature_names, show=['feature_importances']) 
```

```python
eli5.show_prediction(rfc, X_test.iloc[0])
```
